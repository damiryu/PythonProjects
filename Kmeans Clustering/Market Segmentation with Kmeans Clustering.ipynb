```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('3.12.+Example.csv')
data
```

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

```python
x = data.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
```

```python
clusters['predicted_clusters'] = kmeans.fit_predict(clusters)
clusters
```

```python
plt.scatter(clusters['Satisfaction'], clusters['Loyalty'], c=clusters['predicted_clusters'], cmap='rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```

From the graph above, it may look as though the clustering was done only on the basis of satisfaction. To fix this problem, we have to give both satisfaction and loyalty equal weight since they are equally important.
- This can be done by standardization

Standardize the variable

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

Take advantage of the elbow method to determine the number of clusters required

```python
wcss = []
for i in range(1,10):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
plt.plot(range(1,10), wcss)
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
```

Exploring several number of clusters from the graph

```python
kmeans = KMeans(4)
kmeans.fit(x)
```

```python
clusters_new = x.copy()
clusters_new['new_predicted_clusters'] = kmeans.fit_predict(x_scaled)
clusters_new
```

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c=clusters_new['new_predicted_clusters'], cmap = 'rainbow')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.show()
```
